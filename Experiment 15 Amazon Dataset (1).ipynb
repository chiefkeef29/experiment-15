## Part A - Data Normalisation (Amazon Dataset)

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/Amazon Products Dataset Expt 14.csv")

# Keep relevant numeric columns and rename to match experiment structure
df = df[['name', 'actual_price', 'no_of_ratings', 'discount_percentage']].copy()
df.columns = ['Product', 'Price', 'Units_Sold', 'Discount']

# Clean: remove commas/symbols and convert to numeric
df['Price'] = pd.to_numeric(df['Price'].astype(str).str.replace(',','').str.replace('₹','').str.strip(), errors='coerce')
df['Units_Sold'] = pd.to_numeric(df['Units_Sold'].astype(str).str.replace(',','').str.strip(), errors='coerce')
df['Discount'] = pd.to_numeric(df['Discount'].astype(str).str.replace('%','').str.strip(), errors='coerce')

df = df.dropna().reset_index(drop=True)
print(df.head())
```

```python
df['Price_MinMax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min())
print(df[['Product','Price', 'Price_MinMax']])
df
```

```python
cols = ['Price', 'Units_Sold', 'Discount']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
df_norm
```

```python
df['Units_Zscore'] = (df['Units_Sold'] - df['Units_Sold'].mean()) / df['Units_Sold'].std()
print(df[['Product', 'Units_Sold', 'Units_Zscore']])
```

```python
df['Price_Decimal'] = df['Price'] / df['Price'].max()
print(df[['Product', 'Price', 'Price_Decimal']])
```

```python
df['Units_Decimal'] = df['Units_Sold'] / df['Units_Sold'].max()
print(df[['Product', 'Units_Sold', 'Units_Decimal']])
```

## Part B - Data Encoding (Amazon Dataset)

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("/content/Amazon Products Dataset Expt 14.csv")

# Select relevant categorical and numeric columns
df1 = df1[['name', 'main_category', 'sub_category', 'ratings', 'actual_price']].copy()
df1.columns = ['Product_Name', 'Customer_Gender', 'Payment_Method', 'City', 'Order_Value']

# Clean Order_Value
df1['Order_Value'] = pd.to_numeric(df1['Order_Value'].astype(str).str.replace(',','').str.replace('₹','').str.strip(), errors='coerce')
df1['City'] = df1['City'].astype(str).str.strip()

df1 = df1.dropna().reset_index(drop=True)
print(df1.head())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Gender_Label'] = le.fit_transform(df1['Customer_Gender'])
df1
```

```python
df1['City_Label'] = le.fit_transform(df1['City'])
df1
```

```python
df1_encoded1 = pd.get_dummies(df1, columns = ['Customer_Gender'])
df1_encoded1
```

```python
df1_encoded1 = pd.get_dummies(df1, columns = ['Payment_Method'], drop_first=True)
df1_encoded1
```
